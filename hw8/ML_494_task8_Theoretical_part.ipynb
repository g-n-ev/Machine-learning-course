## Задача 3

Функционал данной задачи: $U = \sum_1^n{argmin_j{||x_i - c_j||_2}}$, где  $c_j, j = {1,K}$ - центры кластеров, $x_i, i = {1,N}$ - точки.
Значение функционала меняется при переотнесении точки к другому кластеру, либо при изменении координаты центра.
* при переотнесении точки $x_0$ к другому кластеру:
значение $argmin_j{||x_0 - c_j||_2)$ - уменьшается(раз мы переотносим точку), а значит и значение функционала уменьшается

* при изменении координат центра $c_0$:
найдем минимум функционала при уловии что,например , $x_1,x_2,...,x_k$ относятся к кластеру с центром $c_0$. Для упрощения, будем считать, что выборка одномерная. Иначе, это можно расширет на многомерное пространство.
$U'_{c_0} = 2*\sum_1^k{(c_0 - x_i)} = 0$ Следовательно, $c_0^{new} = \frac{\sum_1^k {x_i}}{k}$, то есть перенос в центра кластера в его центр масс приводит к минимизации функционала.

Таким образом, оба действия уменьшают значение функционала, значение которого неотрицательное. А значит, функционал сходится.

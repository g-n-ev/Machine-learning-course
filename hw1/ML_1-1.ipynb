```python
import numpy as np
from matplotlib.colors import ListedColormap
from sklearn import cross_validation, datasets, metrics, neighbors 
%pylab inline
```

```python
estimator = neighbors.KNeighborsClassifier(n_neighbors=1)
```

```python
classification_problem = datasets.make_classification(n_samples=500, n_features =2, n_informative = 2, 
                                                      n_classes = 4, n_redundant=0, 
                                                      n_clusters_per_class=1, random_state=42)
```

```python
classification_problem[0][:20], classification_problem[1][:20]
```

```python
colors = ListedColormap(['b', 'g', 'r','y'])
light_colors = ListedColormap(['lightcoral', 'lightblue', 'lightyellow', 'lightgreen'])
```

```python
pylab.figure(figsize=(10,10))
pylab.scatter(map(lambda x: x[0], classification_problem[0]), map(lambda x: x[1], classification_problem[0]),  c=classification_problem[1],cmap=colors, s=20)
```

```python
train_data, test_data, train_labels, test_labels = cross_validation.train_test_split(classification_problem[0], 
                                                                                     classification_problem[1], 
                                                                                     test_size = 0.3,
                                                                                     random_state = 1)
```

```python
clf = neighbors.KNeighborsClassifier()
clf.fit(train_data, train_labels)
```

```python
def get_meshgrid(data, step=.05, border=.5,):
    x_min, x_max = data[:, 0].min() - border, data[:, 0].max() + border
    y_min, y_max = data[:, 1].min() - border, data[:, 1].max() + border
    return np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))
```

## n = 1

```python
estimator = neighbors.KNeighborsClassifier(n_neighbors=1)
estimator.fit(train_data, train_labels)
    
pyplot.figure(figsize = (10, 10))
    
#plot decision surface on the train data 
pyplot.subplot(1,2,1)
xx, yy = get_meshgrid(train_data)
mesh_predictions = np.array(estimator.predict(np.c_[xx.ravel(), yy.ravel()])).reshape(xx.shape)
pyplot.pcolormesh(xx, yy, mesh_predictions, cmap = light_colors)
pyplot.scatter(train_data[:, 0], train_data[:, 1], c = train_labels, s = 15, cmap = colors)
pyplot.title('Train data, accuracy={:.2f}'.format(metrics.accuracy_score(train_labels, estimator.predict(train_data))))

#plot decision surface on the test data
pyplot.subplot(1,2,2)
pyplot.pcolormesh(xx, yy, mesh_predictions, cmap = light_colors)
pyplot.scatter(test_data[:, 0], test_data[:, 1], c = test_labels, s = 25, cmap = colors)
pyplot.title('Test data, accuracy={:.2f}'.format(metrics.accuracy_score(test_labels, estimator.predict(test_data))))
```

```python
cv_score = dict()
for k in range(1,400):
    estimator = neighbors.KNeighborsClassifier(n_neighbors=k)
    cv = cross_validation.cross_val_score(estimator, classification_problem[0], classification_problem[1], scoring='accuracy', cv = 5)
    cv_score[k] =  cv.mean()
```

```python
plt.plot(cv_score.keys(), cv_score.values())
plt.xlabel('n_neighbours')
plt.ylabel('accuracy')
```
